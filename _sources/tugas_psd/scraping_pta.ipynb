## Scraping pta.trunojoyo.ac.id



```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv

hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
def scrap_pta(hal):
  global hades
  a=0
  data=[]
  for page in range(1,hal):
        url = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{page}"
        request = req.get(url,hades).text
        soup = bs(request, 'html.parser')
        prodi = soup.find_all('div',{'id':'begin'})
        for pro in prodi:
          prod = pro.find('h2').text
          jur = prod[-18:]
          utama = soup.find('ul', 'items list_style')
          perid = utama.find_all('li', {'data-id':'id-1'})
          for x in perid:
            link = x.find('a','gray button')['href']
            request2 = req.get(link, hades).text
            soup2 = bs(request2, 'html.parser')
            abst= soup2.find('p',{'align':'justify'}).text.replace('\r','').replace('\n','')
            NPM = x.find('a','gray button')['href'][-12:]
            headline = x.find('a', 'title').text.replace('\r','').replace('\n','')
            print(f'done[{a}] > {headline[0:10]}')
            a+=1
            data.append([NPM,headline,abst,jur])
            kepala = ['NPM','Judul','Abstrak']
            writer = csv.writer(open('/content/drive/MyDrive/Prosaindata/tugas/tugas_psd/pta.csv', 'w', newline=''))
            writer.writerow(kepala)
            for datas in data :
              writer.writerow(datas)
```

```python
scrap_pta(172)
```
