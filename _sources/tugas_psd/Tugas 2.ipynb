# Tugas 2 Pipelines

Untuk mengerjakan tugas Papelines ini, hal yang harus lakukan diantaranya :
1.   Login terlebih dahulu ke web https://cloud.orchest.io/, kemudian buat project dan membuat pipeline per stepnya dan saling terhubung antar step untuk memperoleh hasil akhir akurasi decesion tree dan svm
2.   Kemudian akses https://www.cloudclusters.io/ untuk mengonlinekan data yang berada di lokal.
3.   Step ambil data postgresql elephant dengan menggunakan script yang sudah diberikan selama perkuliahan.
4.   Aktifkan postgresql pada web di no.2 untuk mendapatkan host, port, dll. Buka aplikasi postgresql untuk koneksi hasil aktivasi di web tersebut dan langsung upload file csv. Masukkan script untuk mengakses ke dalam step pada web no.1 yang digunakan untuk menampung data postgresql manual.
5.   Aktifkan mysql pada web di no.2 untuk mendapatkan host, port, dll. Kemudian masuk ke menu phpMyAdmin dan klik Launch untuk koneksi hasil aktivasi di web tersebut dan langsung upload file csv. Masukkan script untuk mengakses ke dalam step pada web no.1 yang digunakan untuk menampung data mysql manual.
6.   Aktifkan mysql server pada web di no.2 untuk mendapatkan host, port, dll. Buka aplikasi ssms untuk koneksi hasil aktivasi di web tersebut dan langsung upload file csv. Masukkan script untuk mengakses ke dalam step pada web no.1 yang digunakan untuk menampung data sql server.
7.   Hasil step yang menampung pengambilan data pada setiap apk digabung menjadi 1 di step gabungan
8. hasil data dari gabungan digunakan untuk mengetahui akurasi data ke jenis klasifikasi decesion tree dan svm
