```python
import networkx as ne
g = ne.Graph()
```

```python
g.add_edges_from([(1,2),(1,3),(1,4),(3,2),(3,4),(4,5),(4,6),(5,6),(5,7),(5,8),(6,7),(6,8),(7,8),(7,9)])
```

```python
ne.draw(g,with_labels=True)
```

```python
# banyaknya node yang terhubung dalam node tertentu
Dc=ne.degree_centrality(g)
Dc
```

```python
# jarak terdekat dari node A ke node G
Cc=ne.closeness_centrality(g)
Cc
```

```python
# deteksi kelas pemisah yang terhubung ke aktor utama
# BC PALING TINGGI MERUPAKAN NODE PALING PENTING
Bc=ne.betweenness_centrality(g)
Bc
```

```python
Pc=ne.pagerank(g)
Pc
```
