```python
import twint
import nest_asyncio
import pandas as pd
```

```python
nest_asyncio.apply()
c = twint.Config()
c.Search = '#prabowo'
c.Pandas = True
c.Limit = 500
twint.run.Search(c)
```

```python
def column_names():
  return twint.output.panda.Tweets_df.columns

def twint_to_pd(columns):
  return twint.output.panda.Tweets_df[columns]

print(column_names())
```

```python
data = twint_to_pd(['username', 'tweet'])
print(data)

data.to_csv('/content/drive/MyDrive/Prosaindata/tugas/tugas_psd/dataprabowo.csv', index=False)
```
